```python
import PyPDF2
import os

path = "C:/Users/Ranja.Sarkar/Desktop/new3"
documents = os.listdir(path)

p2 = []
p1 = []
for document in documents:
    if document.endswith(".pdf"):
                
        in_file = os.path.abspath(path + "\\" + document)
        pdfileobject = open(in_file,'rb')
        pdfreader = PyPDF2.PdfFileReader(pdfileobject)   
        count = pdfreader.numPages
        for i in range(count):
             page = pdfreader.getPage(i)
             
             pp = page.extractText()
             
             p1.append(pp)
        p2.append(p1)
pagetxt = ''.join([str(elem) for elem in p2])

#print(len(documents))
```

```python
from gensim.summarization import keywords

values = keywords(text = pagetxt, split='\n', scores = True)
```

```python
#print(values)
```

```python
import pandas as pd

data = pd.DataFrame(values,columns=['Keyword','score'])
data = data.sort_values('score', ascending = False)
data
```
