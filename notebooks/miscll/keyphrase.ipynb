```python
import PyPDF2
import os

path = "C:/Users/Ranja.Sarkar/Desktop/new3"
documents = os.listdir(path)

p2 = []
p1 = []
for document in documents:
    if document.endswith(".pdf"):
                
        in_file = os.path.abspath(path + "\\" + document)
        pdfileobject = open(in_file,'rb')
        pdfreader = PyPDF2.PdfFileReader(pdfileobject)   
        count = pdfreader.numPages
        for i in range(count):
             page = pdfreader.getPage(i)
             
             pp = page.extractText()
             
             p1.append(pp)
        p2.append(p1)
pagetxt = ''.join([str(elem) for elem in p2])

#print(len(documents))
```

```python
#import nltk
import re
from rake_nltk import Rake, Metric

#r = Rake()
#r = Rake(stopwords=<list of stopwords>, punctuations=<string of puntuations to ignore>)

#pagetxt = re.sub(r'[0-9]+', '', pagetxt)
pagetxt = pagetxt.lower()

#replace none alphanumeric characters (punctuations) in text with spaces
pagetxt = re.sub(r'[^a-zA-Z0-9\s]', '', pagetxt)

#pagetxt = " ".join(pagetxt.split())

r = Rake(
    stopwords = set([
        'a', 'an', 'and', 'are', 'as', 'at', 'be', 'but', 'by','for', 'if', 'in', 'into', 'is', 'it',
       'no', 'not', 'of', 'on', 'or', 'such','that', 'the', 'their', 'then', 'there', 'these',
        'they', 'this', 'to', 'was', 'will', 'with'])
)

#punctuations = set([',','_','=',';','.','!',"'"])
                   
#collocations: bigrams and trigrams
#r = Rake(ranking_metric = Metric.DEGREE_TO_FREQUENCY_RATIO, min_length = 1, max_length = 1)

r = Rake(ranking_metric = Metric.WORD_FREQUENCY, min_length = 1, max_length = 1)
#r = Rake(ranking_metric = Metric.WORD_DEGREE, min_length = 3, max_length = 3)
 
words = r.extract_keywords_from_text(pagetxt)
#phrases = r.get_ranked_phrases()
phrases = r.get_ranked_phrases_with_scores()
phrases
#len(phrases)
```

```python
import pandas as pd

table = pd.DataFrame(phrases,columns=['score','Keyphrases'])
table = table.sort_values('score',ascending=False)
#table.to_csv('sample1.csv')
```
