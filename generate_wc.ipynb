```python
import pandas as pd
import numpy as np 
import os
import PyPDF2
import string

from wordcloud import WordCloud
import matplotlib.pyplot as plt
from PIL import Image
#from wordcloud import ImageColorGenerator
```

```python

path = "C:/Users/Ranja.Sarkar/Desktop/articles"
docs = os.listdir(path)
for index, i in  enumerate(docs):
    print(index, i)
```

```python
# remove punctuations from all docs 

exclude = set(string.punctuation)
alldocslist = []

for index, i in  enumerate(docs):
    doctext = []
    if i.endswith(".pdf"):
        in_file = os.path.abspath(path + "\\" + i)
        in_file1 = in_file.encode('UTF-8')
        pdfileobject = open(in_file1,'rb')
        pdfreader = PyPDF2.PdfFileReader(pdfileobject)   
        count = pdfreader.numPages
        for j in range(count):
            page = pdfreader.getPage(j)
            pp = page.extractText()
            doctext.append(pp)
        doctext = ''.join(ch for ch in doctext if ch not in exclude)
                             
        alldocslist.append(doctext)  

len(alldocslist)
```

```python
wc = WordCloud(background_color = "white", max_words = 1000, width = 600, height = 600, random_state = 1).generate(doctext)
```

```python
## TF (word cloud) is good for text similarity, in general

plt.imshow(wc)
plt.axis('off')
#plt.imshow(wc.recolor(color_func = image_colors))
plt.savefig('wc.png', figsize = (1200, 900), dpi = 500)
```
