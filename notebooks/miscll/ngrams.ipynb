```python
import PyPDF2
import os

path = "C:/Users/Ranja.Sarkar/Desktop/new4"
documents = os.listdir(path)

p2 = []
p1 = []
for document in documents:
    if document.endswith(".pdf"):
                
        in_file = os.path.abspath(path + "\\" + document)
        pdfileobject = open(in_file,'rb')
        pdfreader = PyPDF2.PdfFileReader(pdfileobject)   
        count = pdfreader.numPages
        for i in range(count):
             page = pdfreader.getPage(i)
             
             pp = page.extractText()
             
             p1.append(pp)
        p2.append(p1)
pagetxt = ''.join([str(elem) for elem in p2])

#print(len(documents))
print(in_file)
```

```python
import re
#import nltk
from nltk.util import ngrams
#nltk.download('stopwords')
#from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

pagetxt = pagetxt.lower()

#Replace all none alphanumeric characters with spaces
pagetxt = re.sub(r'[^a-zA-Z0-9\s]', ' ', pagetxt)

pagetxt = re.sub(r'[0-9]+', '', pagetxt)

stopwords = set([
        'a', 'an', 'and', 'are', 'as', 'at', 'be', 'but', 'by',
        'for', 'if', 'in', 'into', 'is', 'it',
        'no', 'not', 'of', 'on', 'or', 's', 'such',
        't', 'that', 'the', 'their', 'then', 'there', 'these',
        'they', 'this', 'to', 'was', 'will', 'with'
    ])

#stop_words = set(stopwords.words("english"))

#bag of words
#tokens = [token for token in pagetxt.split(" ") if token != ""]

tokens = word_tokenize(pagetxt)
filtered_tokens= [w for w in tokens if not w in stopwords]
```

```python
tri_grams = list(ngrams(filtered_tokens, 3))
#print(out_grams)
print(len(tri_grams))
print(tri_grams)
```

```python
grams = [' '.join(tri_grams[n]) for n in range(len(tri_grams))]
```
